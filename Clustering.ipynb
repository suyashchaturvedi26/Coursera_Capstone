```python
import pandas as pd
import numpy as np

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)

title = soup.title
print(title)
```

```python
rows = soup.find_all('tr')

list_rows = []

for row in rows:
    row_td = row.find_all('td')
    str_cells = str(row_td)
    cleantext = BeautifulSoup(str_cells, "lxml").get_text()
    list_rows.append(cleantext)

df = pd.DataFrame(list_rows)
df1 = df[0].str.split(',',2, expand=True)
```

```python
df1[0] = df1[0].str.strip('[ ] \n')
df1[1] = df1[1].str.strip('[ ] \n')
df1[2] = df1[2].str.strip('[ ] \n')
df1=df1[:181]
df1.drop(0,inplace=True)
```

```python
df1=df1[df1[1]!='Not assigned']

col_labels = soup.find_all('th')

all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
df2=pd.DataFrame(all_header)
```

```python
header = df2[0].str.split(',',expand=True)
header[0] = header[0].str.strip('[ ] \n')
header[1] = header[1].str.strip('[ ] \n')
header[2] = header[2].str.strip('[ ] \n')
header.drop(3,axis=1,inplace=True)
```

```python
frames = [header, df1]
df4 = pd.concat(frames)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.drop(0,inplace=True)
df5.reset_index(drop=True,inplace=True)
```

```python
i = 0
for i in range(0,df5.shape[0]):
    if df5.iloc[i][2] == 'Not assigned':
        df5.iloc[i][2] = df5.iloc[i][1]
        i = i+1
```
